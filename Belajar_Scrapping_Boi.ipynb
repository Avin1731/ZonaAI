```python
import requests
from bs4 import BeautifulSoup
from collections import Counter

def scan_all_elements(url):
    print(f"üïµÔ∏è‚Äç‚ôÇÔ∏è  Sedang membedah TOTAL struktur: {url} ...\n")

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    try:
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"‚ùå Gagal akses. Status: {response.status_code}")
            return

        soup = BeautifulSoup(response.content, 'html.parser')

        # 1. AMBIL SEMUA TAG TANPA KECUALI
        all_tags = [tag.name for tag in soup.find_all()]

        # 2. Hitung statistik tag (Contoh: div ada 100, a ada 50)
        tag_counts = Counter(all_tags)

        # Urutkan dari yang paling sering muncul
        sorted_tags = tag_counts.most_common()

        print(f"üìä Ditemukan {len(sorted_tags)} jenis elemen HTML berbeda.")
        print("="*70)

        # 3. Loop setiap jenis tag yang ditemukan
        for tag_name, count in sorted_tags:
            print(f"üí† Tag: <{tag_name}> | Total: {count}")

            # Ambil semua elemen dari tag jenis ini
            elements = soup.find_all(tag_name)

            # Tampilkan maksimal 3 sampel saja biar layar gak penuh
            for i, el in enumerate(elements[:3], 1):

                # --- A. Ambil Info Class & ID (PENTING BUAT SELECTOR) ---
                class_list = el.get('class', [])
                class_str = ".".join(class_list) if class_list else ""
                id_str = f"#{el.get('id')}" if el.get('id') else ""

                selector_info = f"{tag_name}"
                if id_str: selector_info += id_str
                if class_str: selector_info += f".{class_str}"

                # --- B. Bersihkan Text ---
                # Hati-hati, script & style isinya kode, bukan teks bacaan
                if tag_name in ['script', 'style', 'meta', 'link', 'noscript']:
                    text_content = "[System Code/Metadata - Skipped]"
                else:
                    text_content = el.get_text(" ", strip=True)
                    # Potong kalau kepanjangan
                    if len(text_content) > 80: text_content = text_content[:77] + "..."
                    if not text_content: text_content = "[Tidak ada teks / Container kosong]"

                # --- C. Info Tambahan (Link/Image) ---
                extra = ""
                if tag_name == 'a': extra = f" üîó Href: {el.get('href', '')[:30]}..."
                if tag_name == 'img': extra = f" üñºÔ∏è Src: {el.get('src', '')[:30]}..."
                if tag_name == 'meta': extra = f" ‚ÑπÔ∏è Content: {el.get('content', '')[:30]}..."

                # PRINT OUTPUT
                print(f"   Sample {i}: {selector_info}")
                print(f"   ‚Ü≥ Isi: {text_content}{extra}")

            print("-" * 70)

    except Exception as e:
        print(f"Error: {e}")

# --- JALANKAN ---
target_url = 'https://edition.cnn.com/2026/02/10/tech/china-us-ai-race-challenges-intl-hnk-dst'
scan_all_elements(target_url)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_cnn_article(url):
    print(f"üöÄ Memulai scraping berita: {url}\n")

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    try:
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print("Gagal koneksi.")
            return

        soup = BeautifulSoup(response.content, 'html.parser')

        # --- 1. AMBIL JUDUL (HEADLINE) ---
        elemen_judul = soup.find('h1', class_='headline__text')
        judul = elemen_judul.text.strip() if elemen_judul else "Judul tidak ditemukan"

        # --- [BARU] 1.5 AMBIL PENULIS (AUTHOR) ---
        # Strategi: Cek Metadata dulu (lebih akurat), kalau nihil baru cek teks visual
        penulis = "-"

        # Cek 1: Metadata (biasanya ada di <meta name="author">)
        meta_author = soup.find('meta', attrs={'name': 'author'})
        if meta_author:
            penulis = meta_author.get('content')
        else:
            # Cek 2: Visual (biasanya class 'byline__names' di CNN)
            elemen_penulis = soup.find(class_='byline__names')
            if elemen_penulis:
                penulis = elemen_penulis.text.strip()

        # --- 2. AMBIL WAKTU (DATE) ---
        elemen_waktu = soup.find('time')
        waktu = elemen_waktu.text.strip() if elemen_waktu else "-"

        # --- 3. AMBIL LOKASI (CITE) ---
        elemen_lokasi = soup.find('cite')
        lokasi = elemen_lokasi.text.strip() if elemen_lokasi else "-"

        # --- 4. AMBIL GAMBAR UTAMA ---
        elemen_gambar = soup.find('img', class_='image__dam-img')
        link_gambar = elemen_gambar['src'] if elemen_gambar else "-"

        # --- 5. AMBIL ISI BERITA (BODY) ---
        artikel_container = soup.find('article')

        isi_berita_full = ""

        if artikel_container:
            list_paragraf = artikel_container.find_all('p')
            temp_text = []

            for p in list_paragraf:
                teks = p.text.strip()
                if len(teks) > 30:
                    temp_text.append(teks)

            isi_berita_full = "\n\n".join(temp_text)
        else:
            isi_berita_full = "Konten artikel tidak terdeteksi."

        # --- TAMPILKAN HASIL RAPI ---
        print("="*60)
        print(f"JUDUL   : {judul}")
        print(f"PENULIS : {penulis}")  # <-- Ditambahkan disini
        print(f"WAKTU   : {waktu}")
        print(f"LOKASI  : {lokasi}")
        print(f"GAMBAR  : {link_gambar}")
        print("-" * 60)
        print("ISI BERITA:\n")
        print(isi_berita_full)
        print("="*60)

    except Exception as e:
        print(f"Error scraping: {e}")

# --- EKSEKUSI ---
target_url = 'https://edition.cnn.com/2026/02/10/tech/china-us-ai-race-challenges-intl-hnk-dst'
scrape_cnn_article(target_url)
```

Kode Scraper Final KOMPAS (Refactored)

```python
import requests
from bs4 import BeautifulSoup

def scrape_kompas_article(url):
    # Tips: Tambahkan '?page=all' agar kalau beritanya berhalaman-halaman,
    # langsung tampil semua dalam satu halaman.
    if '?page=all' not in url:
        url = url + '?page=all'

    print(f"üöÄ Membedah berita Kompas: {url}\n")

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')

        # --- 1. AMBIL JUDUL ---
        # Berdasarkan scan: h1.read__title
        elemen_judul = soup.find('h1', class_='read__title')
        judul = elemen_judul.text.strip() if elemen_judul else "Judul tidak ditemukan"

        # --- 2. AMBIL PENULIS (AUTHOR) ---
        # Kita ambil dari Metadata agar pasti benar (bukan editor)
        penulis = "-"
        meta_author = soup.find('meta', attrs={'name': 'author'}) # Biasanya "Kompas Cyber Media"
        meta_content_author = soup.find('meta', attrs={'name': 'content_author'}) # Nama jurnalis asli

        if meta_content_author:
            penulis = meta_content_author.get('content')
        elif meta_author:
            penulis = meta_author.get('content')

        # --- 3. AMBIL WAKTU ---
        # Kita pakai metadata published_time (format ISO)
        waktu = "-"
        meta_time = soup.find('meta', attrs={'property': 'article:published_time'})
        if meta_time:
            waktu = meta_time.get('content')
        else:
            # Fallback visual: biasanya di div class 'read__time'
            div_time = soup.find('div', class_='read__time')
            if div_time: waktu = div_time.text.strip()

        # --- 4. AMBIL GAMBAR ---
        # Paling aman ambil dari OG:Image (High Res)
        link_gambar = "-"
        meta_img = soup.find('meta', attrs={'property': 'og:image'})
        if meta_img:
            link_gambar = meta_img.get('content')

        # --- 5. AMBIL ISI BERITA (CLEANING "BACA JUGA") ---
        # Container utama Kompas biasanya: div class="read__content"
        container_berita = soup.find('div', class_='read__content')

        isi_berita_full = ""

        if container_berita:
            # Cari semua paragraf
            paragraphs = container_berita.find_all('p')
            temp_text = []

            for p in paragraphs:
                # FILTER 1: Hapus "Baca juga"
                # Di log kamu terlihat: <strong>Baca juga: ...</strong>
                # Kita cek apakah ada tag strong di dalam p ini?
                if p.find('strong') and "Baca juga" in p.text:
                    continue # Skip paragraf ini, jangan dimasukkan

                # FILTER 2: Hapus Text Link Promo di akhir
                if "Dapatkan update" in p.text:
                    continue

                teks_bersih = p.text.strip()

                if len(teks_bersih) > 20: # Ambil yg panjang saja
                    temp_text.append(teks_bersih)

            isi_berita_full = "\n\n".join(temp_text)
        else:
            isi_berita_full = "Konten tidak ditemukan (cek class container)."

        # --- OUTPUT ---
        print("="*60)
        print(f"üìÑ JUDUL   : {judul}")
        print(f"‚úçÔ∏è  PENULIS : {penulis}")
        print(f"üïí WAKTU   : {waktu}")
        print(f"üñºÔ∏è  GAMBAR  : {link_gambar}")
        print("-" * 60)
        print("üìù ISI BERITA (Sudah dibersihkan):\n")
        print(isi_berita_full)
        print("="*60)

    except Exception as e:
        print(f"Error scraping: {e}")

# --- EKSEKUSI ---
target_url = 'https://tekno.kompas.com/read/2026/02/12/14010047/chatgpt-resmi-jadi-asisten-ai-militer-amerika-serikat'
scrape_kompas_article(target_url)
```

Kode Scraper Final CNBC ID (Refactored)

```python
import requests
from bs4 import BeautifulSoup

def scrape_cnbc_article(url):
    print(f"üöÄ Membedah berita CNBC Indonesia: {url}\n")

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')

        # --- 1. AMBIL JUDUL ---
        # Sesuai scan: h1.detail-title
        elemen_judul = soup.find('h1', class_='detail-title')
        judul = elemen_judul.text.strip() if elemen_judul else "Judul tidak ditemukan"

        # --- 2. AMBIL PENULIS (AUTHOR) ---
        # CNBC/Detik Group biasanya rapi di Metadata
        penulis = "-"
        meta_author = soup.find('meta', attrs={'name': 'author'})
        if meta_author:
            penulis = meta_author.get('content')
        else:
            # Fallback visual jika meta tidak ada
            div_author = soup.find('div', class_='author')
            if div_author: penulis = div_author.text.strip()

        # --- 3. AMBIL WAKTU ---
        waktu = "-"
        # Coba ambil dari meta publish date (standar CNBC)
        meta_time = soup.find('meta', attrs={'property': 'article:published_time'})
        if meta_time:
            waktu = meta_time.get('content')
        else:
            # Fallback visual: biasanya class 'date'
            div_date = soup.find(class_='date')
            if div_date: waktu = div_date.text.strip()

        # --- 4. AMBIL GAMBAR ---
        # Sesuai scan: figure -> img
        link_gambar = "-"
        # Cara paling akurat lewat Meta OG Image (Resolusi tinggi)
        meta_img = soup.find('meta', attrs={'property': 'og:image'})
        if meta_img:
            link_gambar = meta_img.get('content')
        else:
            # Fallback ambil dari tag figure
            figure = soup.find('figure')
            if figure:
                img = figure.find('img')
                if img: link_gambar = img.get('src')

        # --- 5. AMBIL ISI BERITA (CLEANING "PIlihan Redaksi") ---
        # Container utama: <article>
        article_container = soup.find('article')

        isi_berita_full = ""

        if article_container:
            # PENTING: Hapus elemen pengganggu sebelum mengambil teks!
            # Di log kamu ada: <table class="topiksisip"> (Pilihan Redaksi) -> KITA HAPUS
            pengganggu_table = article_container.find('table', class_='topiksisip')
            if pengganggu_table:
                pengganggu_table.decompose() # Hancurkan elemen ini dari memori

            # Hapus juga div sisipan iklan jika ada (biasanya class 'parasysip')
            sisipan_iklan = article_container.find_all('div', class_='parasysip')
            for iklan in sisipan_iklan:
                iklan.decompose()

            # Sekarang ambil paragraf yang tersisa
            list_paragraf = article_container.find_all('p')
            temp_text = []

            for p in list_paragraf:
                teks = p.text.strip()

                # Filter teks sampah umum di CNBC
                if "SCROLL TO RESUME CONTENT" in teks: continue
                if "Simak Video" in teks: continue

                if len(teks) > 20:
                    temp_text.append(teks)

            isi_berita_full = "\n\n".join(temp_text)
        else:
            isi_berita_full = "Konten artikel tidak ditemukan."

        # --- OUTPUT ---
        print("="*60)
        print(f"üìÑ JUDUL   : {judul}")
        print(f"‚úçÔ∏è  PENULIS : {penulis}")
        print(f"üïí WAKTU   : {waktu}")
        print(f"üñºÔ∏è  GAMBAR  : {link_gambar}")
        print("-" * 60)
        print("üìù ISI BERITA (Bersih tanpa Pilihan Redaksi):\n")
        print(isi_berita_full)
        print("="*60)

    except Exception as e:
        print(f"Error scraping: {e}")

# --- EKSEKUSI ---
target_url = 'https://www.cnbcindonesia.com/tech/20260212145611-37-710649/startup-china-tak-terkenal-tiba-tiba-sudah-kalahkan-google'
scrape_cnbc_article(target_url)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import pandas as pd
import time

# --- 1. KONFIGURASI UMUM ---
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

def get_soup(url):
    """Fungsi helper untuk request dan parse HTML"""
    try:
        # Trik khusus Kompas: tambah ?page=all
        if 'kompas.com' in url and '?page=all' not in url:
            url += '?page=all'

        response = requests.get(url, headers=HEADERS, timeout=10)
        response.raise_for_status()
        return BeautifulSoup(response.content, 'html.parser')
    except Exception as e:
        print(f"‚ùå Error akses {url}: {e}")
        return None

# --- 2. MODUL SCRAPER SPESIFIK ---

def parse_cnn(soup, url):
    """Logika khusus CNN"""
    # Title
    h1 = soup.find('h1', class_='headline__text')
    title = h1.text.strip() if h1 else "No Title"

    # Author
    meta_author = soup.find('meta', attrs={'name': 'author'})
    author = meta_author.get('content') if meta_author else "-"
    if author == "-":
        vis_author = soup.find(class_='byline__names')
        if vis_author: author = vis_author.text.strip()

    # Date
    meta_time = soup.find('meta', attrs={'property': 'article:published_time'})
    date = meta_time.get('content') if meta_time else "-"

    # Image
    meta_img = soup.find('meta', attrs={'property': 'og:image'})
    image = meta_img.get('content') if meta_img else "-"

    # Content
    article = soup.find('article')
    content = ""
    if article:
        paragraphs = article.find_all('p')
        # Filter text pendek & sampah
        clean_text = [p.text.strip() for p in paragraphs if len(p.text.strip()) > 30]
        content = "\n\n".join(clean_text)

    return {
        "Source": "CNN",
        "Title": title,
        "Author": author,
        "Date": date,
        "Content": content,
        "Image_URL": image,
        "Original_URL": url
    }

def parse_kompas(soup, url):
    """Logika khusus Kompas"""
    # Title
    h1 = soup.find('h1', class_='read__title')
    title = h1.text.strip() if h1 else "No Title"

    # Author
    meta_author = soup.find('meta', attrs={'name': 'content_author'})
    if not meta_author: meta_author = soup.find('meta', attrs={'name': 'author'})
    author = meta_author.get('content') if meta_author else "-"

    # Date
    meta_time = soup.find('meta', attrs={'property': 'article:published_time'})
    date = meta_time.get('content') if meta_time else "-"

    # Image
    meta_img = soup.find('meta', attrs={'property': 'og:image'})
    image = meta_img.get('content') if meta_img else "-"

    # Content
    container = soup.find('div', class_='read__content')
    content = ""
    if container:
        paragraphs = container.find_all('p')
        clean_text = []
        for p in paragraphs:
            # Filter 'Baca Juga' dan Promo
            if p.find('strong') and "Baca juga" in p.text: continue
            if "Dapatkan update" in p.text: continue

            txt = p.text.strip()
            if len(txt) > 20: clean_text.append(txt)
        content = "\n\n".join(clean_text)

    return {
        "Source": "Kompas",
        "Title": title,
        "Author": author,
        "Date": date,
        "Content": content,
        "Image_URL": image,
        "Original_URL": url
    }

def parse_cnbc(soup, url):
    """Logika khusus CNBC Indonesia"""
    # Title
    h1 = soup.find('h1', class_='detail-title')
    title = h1.text.strip() if h1 else "No Title"

    # Author
    meta_author = soup.find('meta', attrs={'name': 'author'})
    author = meta_author.get('content') if meta_author else "-"

    # Date
    meta_time = soup.find('meta', attrs={'property': 'article:published_time'})
    date = meta_time.get('content') if meta_time else "-"

    # Image
    meta_img = soup.find('meta', attrs={'property': 'og:image'})
    image = meta_img.get('content') if meta_img else "-"

    # Content
    article = soup.find('article')
    content = ""
    if article:
        # Hapus Tabel Sisipan & Iklan
        for trash in article.find_all(['table', 'div'], class_=['topiksisip', 'parasysip']):
            trash.decompose()

        paragraphs = article.find_all('p')
        clean_text = []
        for p in paragraphs:
            txt = p.text.strip()
            if "SCROLL TO RESUME" in txt or "Simak Video" in txt: continue
            if len(txt) > 20: clean_text.append(txt)
        content = "\n\n".join(clean_text)

    return {
        "Source": "CNBC",
        "Title": title,
        "Author": author,
        "Date": date,
        "Content": content,
        "Image_URL": image,
        "Original_URL": url
    }

# --- 3. CONTROLLER UTAMA (Dispatcher) ---

def universal_scraper(list_urls):
    dataset = []

    print(f"üöÄ Memulai scraping {len(list_urls)} artikel...\n")

    for url in list_urls:
        domain = urlparse(url).netloc
        soup = get_soup(url)

        if not soup: continue

        data = None
        print(f"sedang memproses: {domain} ...")

        # Routing Logic: Pilih Parser berdasarkan Domain
        if 'cnn' in domain:
            data = parse_cnn(soup, url)
        elif 'kompas' in domain:
            data = parse_kompas(soup, url)
        elif 'cnbc' in domain:
            data = parse_cnbc(soup, url)
        else:
            print(f"‚ö†Ô∏è Domain tidak dikenali: {domain}")

        if data:
            dataset.append(data)

        # Jeda biar gak dikira serangan DDoS
        time.sleep(1)

    return pd.DataFrame(dataset)

# --- 4. EKSEKUSI ---

# Masukkan URL dari 3 web berbeda disini
target_urls = [
    'https://edition.cnn.com/2026/02/10/tech/china-us-ai-race-challenges-intl-hnk-dst',
    'https://tekno.kompas.com/read/2026/02/12/14010047/chatgpt-resmi-jadi-asisten-ai-militer-amerika-serikat',
    'https://www.cnbcindonesia.com/tech/20260212145611-37-710649/startup-china-tak-terkenal-tiba-tiba-sudah-kalahkan-google'
]

# Jalankan Scraper
df_hasil = universal_scraper(target_urls)

# --- 5. TAMPILKAN HASIL ---
print("\n‚úÖ HASIL SCRAPING (PREVIEW DATASET):")
print("="*80)
# Tampilkan 3 kolom utama saja agar muat di layar
print(df_hasil[['Source', 'Title', 'Author', 'Date']])
print("="*80)

# Opsi: Simpan ke Excel/CSV untuk dipakai Data Science/Analisis
# df_hasil.to_csv('dataset_berita_ai.csv', index=False)
# print("üìÅ Data berhasil disimpan ke 'dataset_berita_ai.csv'")

# Untuk melihat satu contoh konten full
print("\nüîç Contoh Konten Bersih (Row 1 - Kompas):")
print("-" * 40)
print(df_hasil.iloc[1]['Content'][:300] + "...") # Print 300 huruf pertama
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import pandas as pd
import time
import io

# --- 1. CONFIG & HEADERS ---
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

def get_soup(url):
    try:
        if 'kompas.com' in url and '?page=all' not in url:
            url += '?page=all'
        response = requests.get(url, headers=HEADERS, timeout=15)
        # response.raise_for_status() # Optional: skip error check to keep flow running
        if response.status_code == 200:
            return BeautifulSoup(response.content, 'html.parser')
        return None
    except Exception as e:
        print(f"‚ùå Error koneksi ke {url}: {e}")
        return None

# --- 2. FUNGSI PARSER (CNN, KOMPAS, CNBC) ---
# (Menggunakan logika yang sudah kita sempurnakan sebelumnya)

def parse_cnn(soup, url):
    h1 = soup.find('h1', class_='headline__text')
    title = h1.text.strip() if h1 else "No Title"

    meta_author = soup.find('meta', attrs={'name': 'author'})
    author = meta_author.get('content') if meta_author else "-"
    if author == "-":
        vis_author = soup.find(class_='byline__names')
        if vis_author: author = vis_author.text.strip()

    meta_time = soup.find('meta', attrs={'property': 'article:published_time'})
    date = meta_time.get('content') if meta_time else "-"

    article = soup.find('article')
    content = ""
    if article:
        paragraphs = article.find_all('p')
        clean_text = [p.text.strip() for p in paragraphs if len(p.text.strip()) > 30]
        content = "\n\n".join(clean_text)

    return {"Source": "CNN", "Title": title, "Author": author, "Date": date, "Content": content, "Link": url}

def parse_kompas(soup, url):
    h1 = soup.find('h1', class_='read__title')
    title = h1.text.strip() if h1 else "No Title"

    meta_author = soup.find('meta', attrs={'name': 'content_author'})
    if not meta_author: meta_author = soup.find('meta', attrs={'name': 'author'})
    author = meta_author.get('content') if meta_author else "-"

    meta_time = soup.find('meta', attrs={'property': 'article:published_time'})
    date = meta_time.get('content') if meta_time else "-"

    container = soup.find('div', class_='read__content')
    content = ""
    if container:
        paragraphs = container.find_all('p')
        clean_text = []
        for p in paragraphs:
            if p.find('strong') and "Baca juga" in p.text: continue
            if "Dapatkan update" in p.text: continue
            txt = p.text.strip()
            if len(txt) > 20: clean_text.append(txt)
        content = "\n\n".join(clean_text)

    return {"Source": "Kompas", "Title": title, "Author": author, "Date": date, "Content": content, "Link": url}

def parse_cnbc(soup, url):
    h1 = soup.find('h1', class_='detail-title')
    title = h1.text.strip() if h1 else "No Title"

    meta_author = soup.find('meta', attrs={'name': 'author'})
    author = meta_author.get('content') if meta_author else "-"

    meta_time = soup.find('meta', attrs={'property': 'article:published_time'})
    date = meta_time.get('content') if meta_time else "-"

    article = soup.find('article')
    content = ""
    if article:
        for trash in article.find_all(['table', 'div'], class_=['topiksisip', 'parasysip']):
            trash.decompose()
        paragraphs = article.find_all('p')
        clean_text = []
        for p in paragraphs:
            txt = p.text.strip()
            if "SCROLL TO RESUME" in txt or "Simak Video" in txt: continue
            if len(txt) > 20: clean_text.append(txt)
        content = "\n\n".join(clean_text)

    return {"Source": "CNBC", "Title": title, "Author": author, "Date": date, "Content": content, "Link": url}

# --- 3. MAIN CONTROLLER ---

def run_github_scraper(github_raw_url):
    print("üì• Mengunduh CSV dari GitHub...")

    try:
        # Download konten CSV
        response = requests.get(github_raw_url)
        response.raise_for_status()

        # Baca CSV ke Pandas DataFrame
        df_input = pd.read_csv(io.StringIO(response.text))

        # Cek kolom 'Link'
        if 'Link' not in df_input.columns:
            print("‚ùå Kolom 'Link' tidak ditemukan di CSV!")
            return

        urls = df_input['Link'].tolist()
        print(f"‚úÖ Berhasil memuat {len(urls)} link artikel.\n")

        results = []

        # --- LOOP SCRAPING ---
        for i, url in enumerate(urls):
            print(f"[{i+1}/{len(urls)}] Memproses: {url[:50]}...")

            domain = urlparse(url).netloc
            soup = get_soup(url)

            if not soup:
                print("   ‚ö†Ô∏è Gagal mengambil konten.")
                continue

            data = None
            if 'cnn' in domain:
                data = parse_cnn(soup, url)
            elif 'kompas' in domain:
                data = parse_kompas(soup, url)
            elif 'cnbc' in domain:
                data = parse_cnbc(soup, url)
            else:
                print(f"   ‚ö†Ô∏è Domain tidak didukung: {domain}")

            if data:
                results.append(data)

            # Sleep biar aman
            time.sleep(1)

        # --- OUTPUT ---
        df_final = pd.DataFrame(results)

        print("\n" + "="*80)
        print("üéâ PREVIEW HASIL (3 BARIS PERTAMA)")
        print("="*80)
        # Menampilkan 3 baris pertama sesuai request
        print(df_final[['Source', 'Title', 'Author']].head(3).to_string(index=False))
        print("="*80)

        # Save to CSV
        filename = "hasil_scraping_zona_ai.csv"
        df_final.to_csv(filename, index=False)
        print(f"\nüíæ File lengkap tersimpan sebagai: {filename}")

    except Exception as e:
        print(f"Critical Error: {e}")

# --- EKSEKUSI ---
# Link RAW GitHub (Sudah saya perbaiki dari link blob kamu)
target_csv = "https://raw.githubusercontent.com/Avin1731/ZonaAI/main/Scrapping/Test%20Scrapping%20-%20Sheet1.csv"

run_github_scraper(target_csv)
```
